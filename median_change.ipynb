# Median Change

```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime
import datetime
import statistics
```

```python
# reading in data
data = pd.read_csv('data/gspc.csv')

# dropping unneeded columns
data = data.drop(labels=['High', 'Low', 'Close', 'Volume'], axis=1)

# creating % change column - never was able to read point changes well, always liked % better
data['pct_change'] = (((data['Adj Close'] - data['Open'])/data['Open']) * 100)

# turning date column into date-time object
date = []
for x in data.Date:
    date.append(datetime.datetime.strptime(x, '%Y-%m-%d').date())
       
data.Date = date
```

```python
# asking user how many days they want to look at (oldest entry is from the 1920s)
days = input("How many days do you want to go back? (max is around 35K) ")
days = int(days)
```

```python
# making a plot
fig, ax = plt.subplots(figsize=(15, 7))

# setting up x and y
x = data.Date.tail(days)
y = data['pct_change'].tail(days)

#seperating out gains and losses since an X% gain might be big but an X% loss might be normal
gains = []
losses = []
for change in y:
    if(change > 0): #gain
        gains.append(change)
    else:
        losses.append(change)


# median_change is overall change, other two are for gains and losses
median_change = round(statistics.median(y), 2)
median_gains = round(statistics.median(gains), 2)
median_losses = round(statistics.median(losses), 2)

# plotting change data points based on gain (green) or loss (red)
for date, change in zip(x.to_list(), y.to_list()):
    if(change > 0):
        ax.scatter(date, change, alpha=0.2, color='g');
    else:
        ax.scatter(date, change, alpha=0.2, color='r');
        

ax.set_title('Daily % Change in S&P 500 - last {} Days OR since {}'.format(days, (data.Date.iloc[-1]-datetime.timedelta(days=days))))
ax.set_xlabel('Time')
ax.set_ylabel('% Change')


# horizontal lined to shoe user median gain, median loss, and median change
ax.axhline(median_change, color='black', label= "Median change: {}%".format((round(y.median(), 2))), ls='--');
ax.axhline(median_gains, color='green', label= "Median gain: {}%".format((round(median_gains, 2))), ls='--');
ax.axhline(median_losses, color='red', label= "Median gain: {}%".format((round(median_losses, 2))), ls='--');

# printing out medians for easy reference
print('MEDIAN CHANGE: {}%'.format(median_change))
print('MEDIAN GAIN: {}%'.format(median_gains))
print('MEDIAN LOSS: {}%'.format(median_losses))

# showing the user what the lines mean
ax.legend()

# save fig for GitHub!
plt.savefig('images/sample.png')
```
